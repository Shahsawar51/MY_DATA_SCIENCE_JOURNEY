# **Regularization**

**Regularization** is a technique used in machine learning to prevent a model from overfitting by penalizing complex models with large weights. It adds an additional term to the loss function that discourages large coefficients, helping the model generalize better to unseen data.

Mathematically, the regularized loss function becomes:

Loss = Empirical Error (e.g., MSE) + lambda.Regularization Term






###  1. **What is Expected Value?** (Intuition)

**Expected Value** is the **long-run average** of outcomes of a random process, assuming it is repeated many times.

It tells you:  
> “If I repeat this random event infinitely, what is the average result I would expect?”

It **does not guarantee** that this value will occur in a single trial — it represents a theoretical average over many repetitions.


####  Example: Dice Roll

- Outcomes: 1, 2, 3, 4, 5, 6  
- Each has probability 1\6



###  3. **Expected Value in ML**

In machine learning, we often deal with **randomness** in:

- Data samples (train/test splits),
- Model parameters (due to training),
- Noise in observations.

So, we say that **model predictions** or **loss values** are **random variables**.

Hence, we can write:

- \( E[\hat{f}(x)] \): Expected prediction for input \( x \)
- \( E[L(y, \hat{f}(x))] \): Expected loss (e.g., MSE)

---

### 4. **Why Expected Value Matters in ML**

Expected value gives us a way to **evaluate models theoretically**, by averaging their behavior over all possible training datasets (which we can’t actually observe in practice, but can analyze mathematically).

It’s especially important in understanding:
- **Bias**
- **Variance**
- **Bias-Variance Decomposition**

---

### 5. **Expected Value in Loss Functions**

When we say:
\[
\text{Expected Risk} = E[(Y - \hat{f}(X))^2]
\]
We’re not just taking MSE on one dataset. We’re imagining **all possible datasets**, models trained on them, and predictions they would make — and then taking the average loss.

This theoretical framework helps us design better models and understand generalization.

### 1. **Underfitting Model** (e.g., a simple straight line)
####  Bias: High  
Your model is too simple. Let’s say you're using **a linear model** to fit data that actually has a **non-linear relationship** (e.g., CGPA + behavioral score + interview score). Your model can’t capture the complexity.

So, for **every training set**, it **always gives similar wrong predictions**.

####  Variance: Low  
Since the model is too simple, it doesn't change much if the dataset changes slightly.

####  Total Error: High  
High bias dominates. The model fails consistently.

---

### 2. **Overfitting Model** (e.g., high-degree polynomial)

Now you use a **very flexible model**, like a 15th-degree polynomial that tries to pass through all points in the training set.

####  Bias: Low  
The model captures **every detail**, even noise. So it closely matches the training data.

####  Variance: High  
But if the training data changes slightly, the model shape **changes a lot**, because it's too sensitive to the input.

#### Total Error: High  
Even though the model fits training data well, it fails badly on new data — due to high variance.

---

### 3. **Just Right (Balanced Bias-Variance)**

You choose a **moderately complex model** — for example, a decision tree with limited depth, or a regularized linear regression.

#### Bias: Medium  
It doesn't capture all complexity, but it's generally on target.

####  Variance: Medium  
Doesn’t swing too wildly if data changes a bit.

####  Total Error: Low  
This is the **sweet spot** — generalizes well.
