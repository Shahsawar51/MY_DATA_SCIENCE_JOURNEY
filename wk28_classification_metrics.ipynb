### **Definitions of Classification Metrics**

Classification metrics are quantitative measures used to evaluate the performance of a classification model (like KNN) by comparing its predictions to actual labels. They help assess how well the model distinguishes between classes (e.g., spam vs. not spam, positive vs. negative). The key metrics covered in Week 28 of DSMP 2.0 are:

1. **Accuracy**:
   - **Definition**: The proportion of correct predictions (both true positives and true negatives) out of all predictions made.
   - **Formula**:

     where:
     - TP (True Positives): Correctly predicted positive cases.
     - TN (True Negatives): Correctly predicted negative cases.
     - FP (False Positives): Incorrectly predicted as positive (Type I error).
     - FN (False Negatives): Incorrectly predicted as negative (Type II error).

2. **Precision**:
   - **Definition**: The proportion of true positive predictions out of all positive predictions made by the model. It measures how precise the model is when predicting the positive class.

3. **Recall (Sensitivity or True Positive Rate)**:
   - **Definition**: The proportion of true positive predictions out of all actual positive cases. It measures how well the model identifies positive cases.

4. **F1 Score**:
   - **Definition**: The harmonic mean of precision and recall, balancing both metrics. It’s useful when you want a single score to evaluate a model, especially for imbalanced datasets.
   - **Formula**:  

   image.png
5. **Confusion Matrix**:
   - **Definition**: A table summarizing the model’s predictions by showing counts of TP, TN, FP, and FN. It’s the foundation for calculating other metrics.
   - **Structure** (for binary classification):
     | Actual \ Predicted | Positive | Negative |
     |-------------------|----------|----------|
     | **Positive**      | TP       | FN       |
     | **Negative**      | FP       | TN       |

6. **Type I and Type II Errors**:
   - **Type I Error (False Positive)**: Predicting positive when the actual class is negative (e.g., marking a non-spam email as spam).
   - **Type II Error (False Negative)**: Predicting negative when the actual class is positive (e.g., missing a spam email).

7. **Multi-Class Metrics**:
   - For multi-class problems (e.g., classifying digits 0–9), metrics like precision, recall, and F1 score are calculated per class and then averaged (e.g., macro, micro, or weighted averaging).

### **Key Notes**:
- These metrics are critical for KNN because its performance depends on factors like the choice of `k`, distance metric (e.g., Euclidean, Manhattan), and data quality (e.g., imbalance, as you mentioned in a previous discussion).
- Metrics like precision, recall, and F1 score are especially important when dealing with **imbalanced datasets**, where accuracy alone can be misleading (e.g., if 90% of data is one class, a model predicting that class always can still get 90% accuracy).
