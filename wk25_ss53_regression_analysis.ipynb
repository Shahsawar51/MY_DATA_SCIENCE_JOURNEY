# **Regression Analysis: A Complete Tutorial**

## üìò **1. What is Regression Analysis?**
Regression analysis is a **statistical method** used to understand the relationship between **one or more independent variables (features)** and a **dependent variable (target)**. It helps us:

- **Model relationships** between variables
- **Make predictions** about future outcomes
- **Understand the strength and significance** of predictor variables

### **Example:**
Predicting **house prices** based on features like **area size**, **number of rooms**, and **location**.

---

## üìä **2. Types of Regression**
- **Simple Linear Regression:** One independent variable
- **Multiple Linear Regression:** Multiple independent variables
- **Polynomial Regression:** Non-linear relationships

For now, we‚Äôll focus on **Linear Regression**.

---

## üß† **3. Linear Regression: The Core Equation**

The equation of a **simple linear regression** is:

\[
Y = b_0 + b_1X + \epsilon
\]

Where:
- **Y:** Dependent variable (target)
- **X:** Independent variable (feature)
- **b‚ÇÄ:** Intercept (value of Y when X = 0)
- **b‚ÇÅ:** Slope (change in Y for a unit change in X)
- **œµ:** Error term (residual)

In **multiple linear regression**:

\[
Y = b_0 + b_1X_1 + b_2X_2 + ... + b_nX_n + \epsilon
\]

---

## üìà **4. Measuring Model Fit**

1. **Total Sum of Squares (TSS):** Total variability in the dependent variable.

\[
TSS = \sum (Y_i - \bar{Y})^2
\]

2. **Regression Sum of Squares (RSS):** Variability explained by the model.

\[
RSS = \sum (\hat{Y_i} - \bar{Y})^2
\]

3. **Residual Sum of Squares (ESS):** Unexplained variability (errors).

\[
ESS = \sum (Y_i - \hat{Y_i})^2
\]

These are related as:

\[
TSS = RSS + ESS
\]

---

## ‚úÖ **5. R-Squared (Goodness of Fit)**

**R¬≤** tells us what **proportion of the variance** in the dependent variable is explained by the independent variables:

\[
R^2 = 1 - \frac{ESS}{TSS}
\]

**Interpretation:**
- **R¬≤ = 0.8:** 80% of the variance in Y is explained by X.
- **Higher R¬≤ = Better fit**, but beware of overfitting!

---

## üìä **6. Adjusted R-Squared**

When adding more features, **R¬≤** can increase even if the new features aren‚Äôt significant. **Adjusted R¬≤** corrects this by penalizing unnecessary features:

\[
R_{adj}^2 = 1 - (1 - R^2) \times \frac{n - 1}{n - p - 1}
\]

Where:
- **n:** Number of observations
- **p:** Number of predictors

---

## üîç **7. Hypothesis Testing in Regression**

### **F-Test (Overall Model Significance)**
The **F-statistic** tests whether **at least one coefficient** is non-zero (i.e., at least one feature is useful):

\[
F = \frac{MSR}{MSE} = \frac{RSS/df_{reg}}{ESS/df_{res}}
\]

Where:
- **MSR:** Mean square regression
- **MSE:** Mean square error
- **df_reg:** Degrees of freedom for regression
- **df_res:** Degrees of freedom for residuals

If **p-value < 0.05**, we **reject the null hypothesis** and conclude that **at least one feature affects the target**.

---

## üìå **8. t-Test (Significance of Individual Coefficients)**

The **t-test** checks if **each coefficient** is significantly different from zero:

\[
t = \frac{b_j}{SE(b_j)}
\]

Where:
- **b_j:** Coefficient of predictor **j**
- **SE(b_j):** Standard error of the coefficient

If **p-value < 0.05**, the feature is **significant**.

---

## üîë **9. Confidence Intervals**

A **95% confidence interval** tells us the **range of likely values** for each coefficient. If the interval includes **0**, the coefficient might **not be significant**.

\[
CI = b_j \pm t_{crit} \times SE(b_j)
\]

---

## üîç **10. Assumptions of Linear Regression**

1. **Linearity:** Relationship between X and Y should be linear.
2. **Normality of Residuals:** Errors should follow a normal distribution.
3. **Homoscedasticity:** Constant variance of errors.
4. **No Autocorrelation:** Errors should be independent.
5. **No Multicollinearity:** Features shouldn‚Äôt be too correlated.

If assumptions fail, techniques like **data transformation** or **regularization (Ridge, Lasso)** can help.

---

## üìò **11. Detecting Multicollinearity**

- **Correlation Matrix:** Check feature correlations.
- **Variance Inflation Factor (VIF):** Measures how much the **variance of coefficients** is inflated by multicollinearity.

\[
VIF_j = \frac{1}{1 - R_j^2}
\]

If **VIF > 5 or 10**, multicollinearity is a problem.

### 4. ‚úÖ **Check Assumptions of Linear Regression**  
Linear regression makes certain assumptions, and checking them ensures the model is valid.

Key Assumptions:  
1. **Linearity:** The relationship between the independent and dependent variables is linear.  
2. **Normality of Residuals:** Errors should be normally distributed.  
3. **Homoscedasticity:** Constant variance of residuals.  
4. **No Autocorrelation:** Residuals should not be correlated.

üëâ **Why this step matters:** Violating these assumptions can lead to biased, unreliable results.

---

### 5. üß† **Fit the Linear Model**  
Now that the data is ready and assumptions are checked, it‚Äôs time to fit the regression model.

Linear Regression Formula:  

\[
Y = \beta_0 + \beta_1 X + \epsilon
\]

Where:  
- \( Y \) ‚Üí Dependent Variable  
- \( X \) ‚Üí Independent Variable  
- \( \beta_0 \) ‚Üí Intercept  
- \( \beta_1 \) ‚Üí Slope/Coefficient  
- \( \epsilon \) ‚Üí Error term (residuals)

In Python (using `statsmodels` or `sklearn`):  

```python

```

üëâ **Why this step matters:** This is where the model learns the relationship between X and Y.

- **Intercept:** The value of Y when all X values are zero.  
- **R¬≤ (Coefficient of Determination):** Measures how well the model explains the variance in Y.

üëâ **Why this step matters:** This tells you what the model has learned and whether the findings are meaningful.

Validation techniques:

- **Cross-validation:** Split the data into multiple folds and train/test in cycles.

## üìä **What‚Äôs the Statistic Connection with Regression Analysis?**

Regression analysis is deeply rooted in **statistics** ‚Äî it relies on statistical concepts to model relationships, make predictions, and test the significance of those relationships. Let‚Äôs explore this connection!  

---

### üß† **1. Understanding Relationships (Correlation & Causation)**  
Regression helps us **quantify** the relationship between variables:  
- **Correlation:** Measures the **strength and direction** of a linear relationship between two variables.  
- **Regression:** Goes further by **modeling** the relationship and **predicting** outcomes.

Example:  
If **study hours** and **exam scores** are correlated, regression helps **estimate the score** for a given number of study hours.

üëâ **Statistical Link:** Uses **correlation coefficients** to measure linear relationships.

---

### üèóÔ∏è **2. Regression Coefficients & Hypothesis Testing**  
The **coefficients (Œ≤ values)** in a regression model represent the **impact** of each independent variable on the dependent variable. But how do we know if these effects are **real or just random noise**?  

Statistics helps us answer that through:  
- **p-values:** Test the significance of each coefficient. If **p < 0.05**, the variable has a **statistically significant impact**.  
- **t-tests:** Check if an individual coefficient is **significantly different from zero**.

Example:  
If the coefficient for **study hours** is **5** with a **p-value of 0.01**, it means **1 extra hour of study increases the score by 5 points**, and this effect is **statistically significant**.

üëâ **Statistical Link:** Uses **hypothesis testing (t-test)** to validate coefficients.

---

### üìà **3. Measuring Model Fit (R¬≤ and Adjusted R¬≤)**  
Statistics helps us assess how well the regression model explains the data:  
- **R¬≤ (Coefficient of Determination):** Tells what **percentage of the variation** in the dependent variable is explained by the independent variables.  
- **Adjusted R¬≤:** Adjusts for the **number of predictors**, preventing **overfitting**.

Example:  
If **R¬≤ = 0.85**, the model explains **85% of the variability** in scores based on study hours.

üëâ **Statistical Link:** **R¬≤** comes from the **sum of squares** decomposition in statistics.

---

### ‚ö° **4. Analyzing Variability (TSS, RSS, ESS)**  
Regression splits the total variability in the data into components:  
- **TSS (Total Sum of Squares):** Total variability in the dependent variable.  
- **RSS (Residual Sum of Squares):** Variability not explained by the model (errors).  
- **ESS (Explained Sum of Squares):** Variability explained by the model.

\[
TSS = ESS + RSS
\]

üëâ **Statistical Link:** Uses **sum of squares decomposition** from ANOVA.

---

### üìò **5. Model Significance (F-test and ANOVA)**  
Statistics also helps us test the **overall validity** of the regression model:  
- **F-test:** Compares the **explained variance (ESS)** to the **unexplained variance (RSS)** to check if the model is **better than random guessing**.  
- **ANOVA (Analysis of Variance):** Decomposes the variance and tests whether at least one independent variable has a significant effect on the dependent variable.

Example:  
If the **F-statistic** is large and the **p-value is small**, you can conclude that the regression model is **statistically significant**.

üëâ **Statistical Link:** Uses **F-test** and **ANOVA** to test model validity.

---

### üèÅ **In Summary ‚Äî The Statistical Connection:**  
- üìä **Correlation** ‚Üí Measures linear relationships.  
- üß† **Hypothesis Testing** ‚Üí Validates coefficients (t-test, p-values).  
- üìà **Model Fit** ‚Üí Assessed using **R¬≤ and adjusted R¬≤**.  
- ‚ö° **Variance Analysis** ‚Üí **TSS, RSS, ESS** decompose variability.  
- üìö **Model Significance** ‚Üí **F-test** and **ANOVA** test the model‚Äôs

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate the data
x = 10 * np.random.rand(50)
y = 3 * x - 8 + np.random.randn(50) * 4

# Fit a linear regression model
x = x.reshape(-1, 1)
model = LinearRegression()
model.fit(x, y)

# Calculate the predicted values
y_pred = model.predict(x)

# Plot the scatter plot and regression lines
plt.scatter(x, y, label="Data points")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter plot with increased variability and regression lines")

# Plot the actual population line
x_line = np.linspace(0, 10, 100)
y_actual = 2 * x_line - 5
plt.plot(x_line, y_actual, 'r', label="Population line (m=3, b=-8)")

# Plot the estimated regression line
y_estimated = model.coef_[0] * x_line + model.intercept_
plt.plot(x_line, y_estimated, 'g', label=f"Estimated line (m={model.coef_[0]:.2f}, b={model.intercept_:.2f})")

# Add legend and show the plot
plt.legend()
plt.show()
````

````python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generate synthetic data
np.random.seed(42)
n = 100
x1 = np.random.normal(0, 1, n)
x2 = np.random.normal(0, 1, n)
irrelevant_predictors = np.random.normal(0, 1, (n, 10))

y = 2 * x1 + 3 * x2 + np.random.normal(0, 1, n)

# Helper function to calculate adjusted R-squared
def adjusted_r2(r2, n, k):
    return 1 - (1 - r2) * (n - 1) / (n - k - 1)

# Fit linear regression models with different predictors
X = pd.DataFrame({'x1': x1, 'x2': x2})
X_with_irrelevant = pd.concat([X] + [pd.Series(irrelevant_predictors[:, i], name=f"irrelevant_{i}") for i in range(10)], axis=1)

model1 = LinearRegression().fit(X, y)
model2 = LinearRegression().fit(X_with_irrelevant, y)

# Calculate R-squared and adjusted R-squared for each model
models = [('Model with relevant predictors', model1, X.shape[1]), ('Model with irrelevant predictors', model2, X_with_irrelevant.shape[1])]

for name, model, k in models:
    r2 = r2_score(y, model.predict(X_with_irrelevant.iloc[:, :k]))
    adj_r2 = adjusted_r2(r2, n, k)
    print(f"{name}: R-squared = {r2:.3f}, Adjusted R-squared = {adj_r2:.3f}")
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 7, 8])

# Fit regression model
model = LinearRegression()
model.fit(X, y)

# Predicted values
y_pred = model.predict(X)

# Plot the data points
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Fitted Regression Line')

# Highlight two points that 'fix' the line
plt.scatter([X[0], X[-1]], [y[0], y[-1]], color='green', s=100, label='Fixed Points (for line fitting)')

# Add labels
plt.xlabel('X (Input)')
plt.ylabel('Y (Output)')
plt.title('Regression Line with Fixed Points')
plt.legend()
plt.grid()
plt.show()

# Print slope and intercept
print("Slope (Œ≤‚ÇÅ):", model.coef_[0])
print("Intercept (Œ≤‚ÇÄ):", model.intercept_)

# This code will show you how 2 points essentially fix the regression line!
# Run it in Google Colab and play around with the data points. üöÄ
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example with 100 data points
np.random.seed(42)
X = np.linspace(1, 100, 100).reshape(-1, 1)
y = 3 * X.squeeze() + np.random.normal(0, 15, size=100)

# Fit regression model
model = LinearRegression()
model.fit(X, y)

# Predicted values
y_pred = model.predict(X)

# Plot the data points
plt.scatter(X, y, color='blue', label='Data Points', alpha=0.6)
plt.plot(X, y_pred, color='red', label='Fitted Regression Line')

# Highlight two points that 'fix' the line
plt.scatter([X[0], X[-1]], [y[0], y[-1]], color='green', s=100, label='Fixed Points (for line fitting)')

# Add labels
plt.xlabel('X (Input)')
plt.ylabel('Y (Output)')
plt.title('Regression Line with 100 Data Points')
plt.legend()
plt.grid()
plt.show()

# Print slope and intercept
print("Slope (Œ≤‚ÇÅ):", model.coef_[0])
print("Intercept (Œ≤‚ÇÄ):", model.intercept_)
print("Degrees of Freedom (DOF):", len(X) - 2)

# This visualization shows how 2 points fix the regression line, even with 100 data points! üöÄ
````

````python
import numpy as np
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression

# Example with 100 data points and 2 features
np.random.seed(42)
X1 = np.linspace(1, 100, 100)
X2 = np.random.uniform(1, 50, 100)
y = 3 * X1 + 2 * X2 + np.random.normal(0, 15, size=100)

# Combine features
X = np.column_stack((X1, X2))

# Fit regression model
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

# Create regression plane
x1_range = np.linspace(min(X1), max(X1), 10)
x2_range = np.linspace(min(X2), max(X2), 10)

X1_grid, X2_grid = np.meshgrid(x1_range, x2_range)
y_grid = model.intercept_ + model.coef_[0] * X1_grid + model.coef_[1] * X2_grid

# Create plotly figure
fig = go.Figure()

# Add scatter plot for data points
fig.add_trace(go.Scatter3d(x=X1, y=X2, z=y, mode='markers', marker=dict(size=5, color='blue'), name='Data Points'))

# Add regression plane
fig.add_trace(go.Surface(x=X1_grid, y=X2_grid, z=y_grid, colorscale='reds', opacity=0.5, name='Regression Plane'))

# Update layout
fig.update_layout(
    scene=dict(
        xaxis_title='Feature 1 (X1)',
        yaxis_title='Feature 2 (X2)',
        zaxis_title='Target (Y)'
    ),
    title='3D Regression with 2 Features (Plotly)',
)

fig.show()

# Print coefficients and DOF
print("Slope (Œ≤‚ÇÅ for X1):", model.coef_[0])
print("Slope (Œ≤‚ÇÇ for X2):", model.coef_[1])
print("Intercept (Œ≤‚ÇÄ):", model.intercept_)
print("Degrees of Freedom (DOF):", len(X) - 3)

# Now you can rotate, zoom, and explore the regression plane interactively! üöÄ
````
